***Условие***

Предъявить доверительный интервал уровня $1-α$ для указанного парметра при данных предположениях (с математическими обоснованиями). Сгенерировать 2 выборки объёма 25 и посчитать доверительный интервал. Повторить 1000 раз. Посчитать, сколько раз 95-процентный доверительный интервал покрывает реальное значение параметра. То же самое сделать для объема выборки 10000. Как изменился результат? Как объяснить? Что изменяется при росте объемов выборок?

Даны две независимые выборки $X_1, X_2$ из нормальных распределений $N (\mu_1, σ^2_1)$, $N(\mu_2, σ^2_2)$ объемов $n_1, n_2$ соответственно. Сначала указана оцениваемая функция, потом данные об остальных параметрах, затем параметры эксперимента и подсказки.

$$τ=\frac{σ^2_1}{\sigma^2_2}$$
**Известны** $\mu_1 = \mu_2 = 0$;

$$σ^2_1 = 2, \, \, σ^2_2=1$$
Воспользоваться функцией
$$\frac{n_2 \sum \limits_{i=1}^{n_1} \left( X_{1,i}-\mu_1\right)^2}{n_1 \sum \limits_{i=1}^{n_2} \left( X_{2,i}-\mu_2\right)^2}$$
Смотреть в сторону распределения *Фишера*.



***
***Решение***

***

*   Запишем определение распределения Фишера.

Пусть $Y_1$ и $Y_2 -$ две независимые случайные величины, имеющие распределение хи-квадрат: $Y_i \sim \chi^2(d_i),$ где $d_i \in \mathbb{N}$, $i=1, 2$. Тогда распределение случайной величины $F=\frac{Y_1 / d_1}{Y_2 / d_2}$ называется *распределнием Фишера* со степенями свободы $d_1$ и $d_2$.

*   Также отметим определение для распределения хи-квадрат.

Пусть $z_1,...,z_k-$ совместно независимые стандартные нормальные случайные величины. Тогда случайная величина $x = z_1^2+...+z_k^2$ имеет распределение хи-квадрат с $k$ степенями свободы $x \sim \chi^2(k)$.

*   Приведем нормально распределенные величины к стандартному нормальному распределению
$$\frac{\bar{X}_1 - \mu_1}{σ_1}  \sim Z$$

$$\frac{\bar{X}_2 - \mu_2}{σ_2}  \sim Z$$

*   Теперь воспользуемся определением для хи-распределения.

$$∑\limits_{i=1}^{n_1} \left(\frac{X_{1,i} - \mu_1}{σ_1} \right)^2 \sim \chi^2(n_1)$$

$$∑\limits_{i=1}^{n_2} \left(\frac{X_{2,i} - \mu_2}{σ_2} \right)^2 \sim \chi^2(n_2)$$

*   Запишем распределение Фишера:

$$\frac{Y_1 / d_1}{Y_2 / d_2} = \frac{n_2∑\limits_{i=1}^{n_1} \left(\frac{X_{1,i} - \mu_1}{σ_1} \right)^2 }{n_1∑\limits_{i=1}^{n_2} \left(\frac{X_{2,i} - \mu_2}{σ_2} \right)^2} =
\frac{n_2σ_2^2∑\limits_{i=1}^{n_1} \left(X_{1,i} - \mu_1 \right)^2 }{n_1σ_1^2∑\limits_{i=1}^{n_2} \left(X_{2,i} - \mu_2 \right)^2} = \frac{1}{τ}\frac{n_2∑\limits_{i=1}^{n_1} \left(X_{1,i} - \mu_1 \right)^2 }{n_1∑\limits_{i=1}^{n_2} \left(X_{2,i} - \mu_2 \right)^2} \sim F(n_1, n_2)$$

Для удобства работы с $\tau$ перепишем:
$$f(n_2, n_1)=\tau \frac{n_1∑\limits_{i=1}^{n_2} \left(X_{2,i} - \mu_2 \right)^2 }{n_2∑\limits_{i=1}^{n_1} \left(X_{1,i} - \mu_1 \right)^2} \sim F(n_2, n_1)$$

*   Перейдем к нахождению доверительного интервала.
$$P \left( ϕ_{α /2} <  f(n_2, n_1) < ϕ_{1 -α /2}\right) = 1 - α$$

$$P \left( ϕ_{α /2} <  \tau \frac{n_1∑\limits_{i=1}^{n_2} \left(X_{2,i} - \mu_2 \right)^2 }{n_2∑\limits_{i=1}^{n_1} \left(X_{1,i} - \mu_1 \right)^2} < ϕ_{1 -α /2}\right) = 1 - α$$

$$P \left( ϕ_{α /2}\frac{n_2∑\limits_{i=1}^{n_1} \left(X_{1,i} - \mu_1 \right)^2 }{n_1∑\limits_{i=1}^{n_2} \left(X_{2,i} - \mu_2 \right)^2} <  \tau  < ϕ_{1 -α /2}\frac{n_2∑\limits_{i=1}^{n_1} \left(X_{1,i} - \mu_1 \right)^2 }{n_1∑\limits_{i=1}^{n_2} \left(X_{2,i} - \mu_2 \right)^2}\right) = 1 - α$$

Получаем доверительный интервал для $\tau$:

$$\tau \in  \left( ϕ_{α /2}\frac{n_2∑\limits_{i=1}^{n_1} \left(X_{1,i} - \mu_1 \right)^2 }{n_1∑\limits_{i=1}^{n_2} \left(X_{2,i} - \mu_2 \right)^2} ;  ϕ_{1 -α /2}\frac{n_2∑\limits_{i=1}^{n_1} \left(X_{1,i} - \mu_1 \right)^2 }{n_1∑\limits_{i=1}^{n_2} \left(X_{2,i} - \mu_2 \right)^2} \right)$$

*   Перейдем к программному моделированию.

```python
import numpy as np
import scipy.stats as stats


n_1 = 25
mean_1 = 0
mean_2 = 0
s_1 = 2
s_2 = 1

a = 0.05

t = s_1 / s_2

x_1 = np.random.normal(mean_1, np.sqrt(s_1), n_1)
x_2 = np.random.normal(mean_2, np.sqrt(s_2), n_1)

f_quant_1 = stats.f.ppf(a / 2, n_1, n_1)
f_quant_2 = stats.f.ppf(1 - a / 2, n_1, n_1)


l = f_quant_1 * n_1 * np.sum((x_1 - mean_1) ** 2) / (n_1 * np.sum((x_2 - mean_2) ** 2))
r = f_quant_2 * n_1 * np.sum((x_1 - mean_1) ** 2) / (n_1 * np.sum((x_2 - mean_2) ** 2))

print('Результаты для n =\u001b[34;1m', n_1, '\u001b[0m')
print('-------------------------------------------------------')
print('\u001b[36;1mДоверительный интервал:\u001b[34;1m', '(', l, ';', r,')\u001b[0m')
print('\u001b[36;1mτ = \u001b[34;1m', t,'\u001b[0m')
print('-------------------------------------------------------')

```

```python
tau_in = 0
iter = 1000

for i in range(1000):
  x_1 = np.random.normal(mean_1, np.sqrt(s_1), n_1)
  x_2 = np.random.normal(mean_2, np.sqrt(s_2), n_1)

  f_quant_1 = stats.f.ppf(a / 2, n_1, n_1)
  f_quant_2 = stats.f.ppf(1 - a / 2, n_1, n_1)


  l = f_quant_1 * n_1 * np.sum((x_1 - mean_1) ** 2) / (n_1 * np.sum((x_2 - mean_2) ** 2))
  r = f_quant_2 * n_1 * np.sum((x_1 - mean_1) ** 2) / (n_1 * np.sum((x_2 - mean_2) ** 2))

  if l < t and t < r:
    tau_in = tau_in + 1

print('Результаты для n = \u001b[34;1m', n_1, '\u001b[0mпри повторении\u001b[34;1m', iter, '\u001b[0mраз.')
print('-------------------------------------------------------')
print('\u001b[36;1mПроцент попаданий значения τ в доверительный интервал = \u001b[34;1m', round(tau_in / iter * 100, 4) ,'% \u001b[0m')
print('-------------------------------------------------------')
```

***
***Решение***

***

*   Запишем функцию распределения Лапласа:

$$F(x) = \cases{\frac{1}{2} e^{\nu (x - \mu)}, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, x \leq \mu,\\
1 - \frac{1}{2} e^{-\nu(x-\mu)}, \, \, \, \, \, x > \mu}$$
где $\nu-$ коэффициент масштаба (в нашем случае $\nu = 1$), $\mu-$ коэффициент сдвига.

*   *Центральная предельная теорема (ЦПТ)*:

сумма достаточно большого количества слабо зависимых случайных величин, имеющих примерно одинаковые масштабы, имеет распределение, близкое к нормальному.

Запишем ЦПТ в форме Леви:

при $n \to \infty$
$$\sqrt{n} \frac{\bar{X}_n - \mu}{σ} \to  N(0, 1),$$

где $\bar{X}_n = \frac{1}{n} \sum \limits_{i = 1}^{n} X_i$

*   Перейдем к доверительному интервалу

$$P \left( - c_{1-α / 2}< \sqrt{n} \frac{\bar{X}_n - \mu}{σ}  <  c_{1-α / 2} \right) = 1 - α$$
Подставим в выражение значение $\sigma = \sqrt{\frac{2}{a^2}} = \sqrt{2}$.

$$P \left( - c_{1-α / 2}< \sqrt{n} \frac{\bar{X}_n - \mu}{\sqrt{2}}  <  c_{1-α / 2} \right) = 1 - α$$

$$ -c_{1-α / 2} \sqrt{\frac{2}{n}} - \bar{X}_n < -\mu < c_{1-α / 2} \sqrt{\frac{2}{n}} - \bar{X}_n$$

$$ c_{1-α / 2} \sqrt{\frac{2}{n}} + \bar{X}_n > \mu > -c_{1-α / 2} \sqrt{\frac{2}{n}} + \bar{X}_n$$

$$\mu \in \left(\bar{X}_n -c_{1-α / 2} \sqrt{\frac{2}{n}};  c_{1-α / 2} \sqrt{\frac{2}{n}} + \bar{X}_n \right)$$


*   Перейдем к программному моделированию.
