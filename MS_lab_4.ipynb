1. Построим линейную модель
$$y = X \beta + \epsilon,$$
где $y -$ вектор наблюдений зависимой переменной, $X-$ матрица регрессора, $\beta -$ вектор коэффициентов, $\epsilon-$ вектор ошибок.

*   Запишем формулу для оценки коэффициентов регрессии с помощью метода наименьших квадратов
$$\hat{\beta} = (X^TX)^{-1}X^Ty$$

*   Запишем формулу для вычисления остаточной дисперсии
$$\hat{\sigma}^2 = \frac{\sum^n_{i=1} \epsilon_i^2}{n-k},$$
где $\epsilon = y - X\hat{\beta}$.

*   Построение доверительных интервалов для коэффициентов

$$(-t_{\alpha / 2, n-k} SEE(\hat{\beta}_i) < \hat{\beta}_i < t_{\alpha / 2, n-k} SEE(\hat{\beta}_i),$$
где $SEE-$ стандартная ошибка регрессии:
$$SEE = \sqrt{\hat{\sigma}^2} = \sqrt{{\frac{1}{n-k} \sum_{i=1}^n \epsilon_i^2}}$$

*   Чем меньше стандартная ошибка регрессии, тем лучше модель соответствует доступным наблюдениям.

*   Коэффициент детерминации $R^2-$ еще одна характеристика качества подгонки
   *   Для множественной регрессии с константой так же, как и для парной, общая сумма квадратов может быть представлена как сумма квадратов остатков и объясненная сумма квадратов:
$$\sum_{i=1}^n (y_i - \overline{y})^2 = \sum_{i=1}^n \epsilon_i^2 + \sum_{i=1}^n (\hat{y}_i - \overline{y})^2$$

   *  $R^2$ рассчитывается по следующим формулам:
   $$R^2 = 1 - \frac{\sum_{i=1}^n \epsilon^2_i}{\sum^n_{i=1} (y_i - \overline{y})^2} = \frac{\sum_{i=1}^n (\hat{y}_i - \overline{y})^2}{\sum_{i=1}^n (y_i - \overline{y})^2} = \frac{\hat{Var}(\hat{y})}{\hat{Var}(y)}$$

```python
import scipy.stats as sps


alpha = 0.05
x = car_data[['MPG.city', 'MPG.highway', 'Horsepower']]
x = np.column_stack((np.ones(x.shape[0]), x))
n, k = x.shape
y = car_data['Price']

beta_ev = np.linalg.inv(x.T @ x) @ x.T @ y

print('\u001b[34;1m Оценка коэффициентов линейной регрессии:\u001b[0m')
print('\u001b[34;1m β_0:\u001b[36;1m', np.round(beta_ev[0],  3), '\u001b[0m')
print('\u001b[34;1m β_1:\u001b[36;1m', np.round(beta_ev[1], 3), '\u001b[0m')
print('\u001b[34;1m β_2:\u001b[36;1m', np.round(beta_ev[2], 3), '\u001b[0m')
print('\u001b[34;1m β_3:\u001b[36;1m', np.round(beta_ev[3], 3), '\u001b[0m')
print()

sig2_ev = ((y - x @ beta_ev).T @ (y - x @ beta_ev)) / (n - k)

print('\u001b[34;1m Оценка остаточной дисперсии:\u001b[0m')
print('\u001b[34;1m σ^2:\u001b[36;1m', np.round(sig2_ev, 3), '\u001b[0m')
print()

t_q = sps.t.ppf(1 - alpha / 2, n - k)
s_er = np.sqrt(np.diag(sig2_ev * np.linalg.inv(x.T @ x)) / (n - k))

print('\u001b[34;1m Доверительные интервалы для коэффициентов:\u001b[0m')
for i in range(len(beta_ev)):
  print('\u001b[34;1m β_',i,':\u001b[36;1m (', np.round(beta_ev[i] - t_q * s_er[i],3),',', np.round(beta_ev[i] + t_q * s_er[i],3),')\u001b[0m')

print()

r_2 = 1 - np.sum((y - x @ beta_ev) ** 2) / np.sum((y - np.mean(y)) ** 2)

print('\u001b[34;1m Коэффициент детерминации:\u001b[0m')
print('\u001b[34;1m R^2:\u001b[36;1m', np.round(r_2, 3), '\u001b[0m')



```

*   Проверка подозрений

   *   Чем больше мощность, тем больше цена

   *   Цена изменяется в зависимости от расхода в городе

   *   Проверить гипотезу $H_0$ о равенстве одновременно нулю коэффициентов при расходе в городе и расходе на шоссе против альтернативы $H_1 = \overline{H_0}$

```python
t_h_p = beta_ev[3] / s_er[3]
p_val_h_p = 1 - sps.t.cdf(t_h_p, n - k)

print('\u001b[34;1m Проверка гипотезы о росте цены при увеличении мощности:\u001b[0m')
print('\u001b[34;1m Значение статистики:\u001b[36;1m', t_h_p, '\u001b[0m')
print('\u001b[34;1m p_value:\u001b[36;1m', p_val_h_p, '\u001b[0m')

if p_val_h_p < alpha:
  print('\u001b[34;1m Гипотеза верна: При увеличении мощности растет и цена\u001b[0m')

print()

t_c_p = beta_ev[1] / s_er[1]
p_val_c_p = 1 - sps.t.cdf(t_c_p, n - k)

print('\u001b[34;1m Проверка гипотезы об изменении цены в зависимости от расходов в городе:\u001b[0m')
print('\u001b[34;1m Значение статистики:\u001b[36;1m', t_c_p, '\u001b[0m')
print('\u001b[34;1m p_value:\u001b[36;1m', p_val_c_p, '\u001b[0m')

if p_val_c_p > alpha:
  print('\u001b[34;1m Гипотеза не верна: нет зависимости изменения цены от расходов в городе\u001b[0m')

print()

f_zero = (np.sum((y - np.mean(y)) ** 2) - np.sum((y - x @ beta_ev) ** 2)) / (np.sum((y - x @ beta_ev) ** 2) / (n - k))
p_val_zero = 1 - sps.f.cdf(f_zero, k - 1, n - k)

print('\u001b[34;1m Проверка гипотезы о равенстве одновременно нулю коэффициентов при расходе в городе и расходе на шоссе:\u001b[0m')
print('\u001b[34;1m Значение статистики:\u001b[36;1m', f_zero, '\u001b[0m')
print('\u001b[34;1m p_value:\u001b[36;1m', p_val_zero, '\u001b[0m')

if p_val_zero < alpha:
  print('\u001b[34;1m Гипотеза не верна: хотя бы один коэффициент не 0\u001b[0m')

print()



```

Для каждого варианта требуется проверить гипотезу о равенстве средних на каждом уровне фактора с помощью модели однофакторного дисперсионного анализа.

*   Сформулируем гипотезы
   *   Нулевая гипотеза $H_0$: Равенство средних выполняется на каждом уровне фактора
   *   Альтернативная гипотеза $H_1$: Равенство средних нарушается хотя бы  на одном уровне фактора

*   Вычисления проводятся по следующим формулам:
   *   Сумма квадратов:
   $$SS_T = \sum_{i=1}^n (y_i - \overline{y})^2$$
   *   Сумма квадратов внутри групп:
   $$SS_W = \sum_{j=1}^m \sum_{i=1}^{n_j} (y_{i,j} - \overline{y}_j)^2$$
   *   Средняя квадратичная ошибка внутри групп:
   $$MS_W = \frac{SS_W}{n - m}$$
   *   Сумма квадратов между группами:
   $$SS_B = \sum_{j=1}^m n_j (\overline{y}_j - \overline{y})^2$$
   *   Средняя квадратичная ошибка между группами:
   $$MS_B = \frac{SS_B}{m - 1}$$
   *   
   $$F = \frac{MS_B}{MS_W}$$

```python
ph_data_pr_bat = phone_data[['price_range', 'battery_power']]
bat_pow = phone_data['battery_power']

y_ = np.mean(bat_pow)
sst = np.sum((bat_pow - y_) ** 2)

gr = ph_data_pr_bat.groupby('price_range')['battery_power'].apply(list)
ssw = np.sum([np.sum((j - np.mean(j)) ** 2) for j in gr])
ssb = np.sum([len(j) * (np.mean(j) - y_) ** 2 for j in gr])

f = (ssb  * (len(ph_data_pr_bat) - len(gr))) / (ssw  * (len(gr) - 1))

print('\u001b[34;1m Значение статистики:\u001b[36;1m', f, '\u001b[0m')

f_cr = sps.f.ppf(1 - alpha, len(gr) - 1, len(ph_data_pr_bat) - len(gr))

print('\u001b[34;1m Критическое значение:\u001b[36;1m', f_cr, '\u001b[0m')

if f > f_cr:
   print('\u001b[34;1m Гипотеза не верна: Равенство средних нарушается хотя бы на одном уровне фактора \u001b[0m')
```
